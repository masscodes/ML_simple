```python
import pandas as pd
pd.set_option('display.max_columns', 500)
from matplotlib import pyplot as plt
```

```python
ratings = pd.read_csv('/content/ratings.csv')
movies = pd.read_csv('/content/movies.csv')
movies.head(50)
```

```python
movies.title = movies.title.map(lambda mov: mov.split('(')[0])
movies.head(1)
```

```python
fav_movies = ['Persuasion ', 'Leaving Las Vegas']
```

```python
movie_ids = movies.loc[movies.title.isin(fav_movies)].movieId
movie_ids
```

```python
ratings = ratings.loc[ratings.rating >= 4]
```

```python

usr = ratings.userId.unique()
users = pd.Series([u for u in usr if set(movie_ids).issubset(set(ratings.loc[ratings.userId == u].movieId))])
```

```python
users.head()
```

```python
mov = ratings.loc[(ratings.userId.isin(users)) & (ratings.rating > 4)]
movg = mov.groupby('movieId').userId.count().nlargest(20)
print(movg)
movies.loc[movies.movieId.isin(movg.index)]
```
